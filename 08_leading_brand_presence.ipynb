```python
# ============================================
# Sephora: Connected Brand Charts (FROM THE TOP)
# Goal: Make bar charts "connected" by using:
#  - One cleaned/reliable dataset
#  - One shared TOP brand list
#  - One shared brand order across ALL plots
#  - One shared summary table feeding all charts
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------
# 0) Load cleaned dataset
# ----------------------------
df = pd.read_csv("/Users/natashamirchevska/Desktop/sephora project /product_info_cleaned.csv", low_memory=False)
print("Initial shape:", df.shape)
display(df.head())

# ----------------------------
# 1) Keep only columns needed
# ----------------------------
use_cols = [
    "product_id",
    "product_name",
    "brand_name",
    "primary_category",
    "price_usd",
    "loves_count",
    "rating",
    "reviews"
]
d = df[use_cols].copy()

# ----------------------------
# 2) Clean numerics + enforce reliability (consistent with your previous requirements)
# ----------------------------
d["price_usd"] = pd.to_numeric(d["price_usd"], errors="coerce")
d["loves_count"] = pd.to_numeric(d["loves_count"], errors="coerce")
d["rating"] = pd.to_numeric(d["rating"], errors="coerce")
d["reviews"] = pd.to_numeric(d["reviews"], errors="coerce")

# Drop rows missing essentials
d = d.dropna(subset=["brand_name", "price_usd", "loves_count", "rating", "reviews"]).copy()

# Reliability: minimum reviews per product
MIN_REVIEWS = 50
d = d[d["reviews"] >= MIN_REVIEWS].copy()

# Optional (uncomment if you want to remove "no engagement" products like before)
d = d[d["loves_count"] > 0].copy()

print("After cleaning + reliability filters:", d.shape)

# ----------------------------
# 3) Anchor list: Top brands by product count (this "connects" all bar charts)
# ----------------------------
TOP_N = 15

brand_counts = (
    d.groupby("brand_name")["product_id"]
     .nunique()
     .sort_values(ascending=False)
     .head(TOP_N)
     .reset_index(name="product_count")
)

# Fixed order used across all charts
brand_order = brand_counts["brand_name"].tolist()

# ----------------------------
# 4) ONE summary table for ALL charts (connected logic)
# ----------------------------
brand_summary = (
    d[d["brand_name"].isin(brand_order)]
    .groupby("brand_name", as_index=False)
    .agg(
        product_count=("product_id", "nunique"),
        median_price=("price_usd", "median"),
        mean_price=("price_usd", "mean"),
        median_rating=("rating", "median"),
        mean_rating=("rating", "mean"),
        median_loves=("loves_count", "median"),
        mean_loves=("loves_count", "mean"),
        median_reviews=("reviews", "median"),
    )
)

# Apply the same order to all visuals
brand_summary["brand_name"] = pd.Categorical(
    brand_summary["brand_name"], categories=brand_order, ordered=True
)
brand_summary = brand_summary.sort_values("brand_name").reset_index(drop=True)

print("Connected summary table used by all plots:")
display(brand_summary)

# ==========================================================
# CONNECTED BAR CHARTS (same brand list + same ordering)
# ==========================================================

# ----------------------------
# Chart A: Top brands by product count (anchor)
# ----------------------------
plt.figure(figsize=(12, 6))
sns.barplot(data=brand_summary, x="product_count", y="brand_name")
plt.title(f"Top {TOP_N} Brands by Product Count (min reviews ≥ {MIN_REVIEWS})")
plt.xlabel("Number of Products")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()

# ----------------------------
# Chart B: Median price for the same brands (same order)
# ----------------------------
plt.figure(figsize=(12, 6))
sns.barplot(data=brand_summary, x="median_price", y="brand_name")
plt.title(f"Median Price (USD) — Same {TOP_N} Brands (Connected Order)")
plt.xlabel("Median Price (USD)")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()

# ----------------------------
# Chart C: Median rating for the same brands (same order)
# ----------------------------
plt.figure(figsize=(12, 6))
sns.barplot(data=brand_summary, x="median_rating", y="brand_name")
plt.title(f"Median Rating — Same {TOP_N} Brands (min reviews ≥ {MIN_REVIEWS})")
plt.xlabel("Median Rating")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()

# ----------------------------
# Chart D: Median loves for the same brands (same order)
# ----------------------------
plt.figure(figsize=(12, 6))
sns.barplot(data=brand_summary, x="median_loves", y="brand_name")
plt.title(f"Median Loves — Same {TOP_N} Brands (Connected Order)")
plt.xlabel("Median Loves Count")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()

# ==========================================================
# OPTIONAL: Final "Leader" synthesis (single score, still connected)
# This is how you *conclude* a leading brand using the same metrics.
# ==========================================================

s = brand_summary.copy()
s["log_median_loves"] = np.log1p(s["median_loves"])

def minmax(x):
    x = x.astype(float)
    return (x - x.min()) / (x.max() - x.min()) if x.max() != x.min() else 0

s["count_n"]  = minmax(s["product_count"])
s["rating_n"] = minmax(s["median_rating"])
s["loves_n"]  = minmax(s["log_median_loves"])

# Weighted score (easy to justify: customer response > sheer catalog size)
s["leader_score"] = 0.45*s["loves_n"] + 0.45*s["rating_n"] + 0.10*s["count_n"]

s = s.sort_values("leader_score", ascending=False).reset_index(drop=True)
leader_brand = s.loc[0, "brand_name"]

plt.figure(figsize=(12, 6))
sns.barplot(data=s, x="leader_score", y="brand_name")
plt.title(f"Leading Brand Score (Connected from Loves + Rating + Count) — Leader: {leader_brand}")
plt.xlabel("Composite Score (normalized)")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()

print("Leader brand (composite score):", leader_brand)
display(s[["brand_name", "product_count", "median_price", "median_rating", "median_loves", "leader_score"]])
```

```python
import numpy as np
import pandas as pd

# Start from df (cleaned products)
d_out = df.copy()

# Ensure numeric
d_out["loves_count"] = pd.to_numeric(d_out["loves_count"], errors="coerce")
d_out["price_usd"] = pd.to_numeric(d_out["price_usd"], errors="coerce")

# Drop unusable rows
d_out = d_out.dropna(subset=["product_id", "loves_count", "price_usd"]).copy()

# Remove zero-loves products (no popularity signal)
d_out = d_out[d_out["loves_count"] > 0].copy()

# --- Create price_band if missing ---
if "price_band" not in d_out.columns:
    d_out["price_band"] = pd.qcut(
        d_out["price_usd"],
        q=5,
        labels=False,
        duplicates="drop"
    ).astype("Int64")
else:
    d_out["price_band"] = pd.to_numeric(d_out["price_band"], errors="coerce")

# Drop rows where price_band still missing (rare, but safe)
d_out = d_out.dropna(subset=["price_band"]).copy()

# Log transform
d_out["log_loves"] = np.log1p(d_out["loves_count"])

# --- IQR outliers per price band (k=1.5) ---
outlier_list = []
for band, g in d_out.groupby("price_band"):
    q1, q3 = g["log_loves"].quantile([0.25, 0.75])
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    outlier_list.append(g[g["log_loves"] > upper].copy())

outliers = pd.concat(outlier_list, ignore_index=True) if outlier_list else pd.DataFrame()

print("Extreme popularity outliers detected:", outliers["product_id"].nunique() if not outliers.empty else 0)
display(outliers.head())
```

```python
top5_secondary = pd.read_csv("top5_products_per_secondary_category_n200.csv")

overlap = (
    top5_secondary.merge(
        outliers,
        on="product_id",
        how="inner",
        suffixes=("_top5", "_outlier")
    )
)

print("Overlapping products (Top-5 secondary AND extreme outliers):", overlap["product_id"].nunique())
display(overlap.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Base scatter ---
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=d,
    x="price_usd",
    y="loves_count",
    color="lightgrey",
    alpha=0.25,
    edgecolor=None,
    label="All products"
)

# --- Extreme popularity outliers ---
sns.scatterplot(
    data=outliers,
    x="price_usd",
    y="loves_count",
    color="crimson",
    alpha=0.8,
    edgecolor=None,
    label="Extreme popularity outliers"
)

# --- Top-5 per secondary category ---
sns.scatterplot(
    data=top5_secondary,
    x="price_usd",
    y="loves_count",
    marker="^",
    s=120,
    color="royalblue",
    edgecolor="black",
    label="Top-5 per secondary category"
)

# --- Overlap: Top-5 AND outlier ---
sns.scatterplot(
    data=overlap,
    x="price_usd_top5",
    y="loves_count_top5",
    marker="*",
    s=250,
    color="purple",
    edgecolor="black",
    label="Top-5 AND extreme outlier"
)

plt.yscale("log")
plt.xlim(0, d["price_usd"].quantile(0.98))

plt.title(
    "Product Popularity vs Price\n"
    "Overlap of Category Leaders and Extreme Popularity Outliers"
)
plt.xlabel("Price (USD)")
plt.ylabel("Loves Count (log scale)")
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
